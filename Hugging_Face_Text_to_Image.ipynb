### `Working with Stable Diffusion parameters`

*   Negative prompting
*   num_inference_steps
*   height
*   weight
*   num_images_per_prompt








```python
def generate_image(pipe, prompt, params):
  img = pipe(prompt, **params).images

  num_images = len(img)
  if num_images>1:
    fig, ax = plt.subplots(nrows=1, ncols=num_images)
    for i in range(num_images):
      ax[i].imshow(img[i]);
      ax[i].axis('off');

  else:
    fig = plt.figure()
    plt.imshow(img[0]);
    plt.axis('off');
  plt.tight_layout()
```

```python
prompt = """animeaesthetic, a fierce warrior with blue hair in a college dress, wielding a glowing katana, standing atop a shattered building, intense amber eyes, sharp jawline, flowing battle cloak with intricate dragon embroidery, sparks and embers swirling in the air, vibrant colors, dynamic action pose,
glowing crescent moon in the background, epic cinematic atmosphere
"""
#Changed hair colour, dress and building in prompt
params = {}
```

```python
generate_image(pipe, prompt, params)
```

```python
#num inference steps
params = {'num_inference_steps': 100}

generate_image(pipe, prompt, params)
```

```python
#height width
params = {'num_inference_steps': 100, 'width': 512, 'height': int(1.5*640)}

generate_image(pipe, prompt, params)
```

```python
#num_images_per_prompt
params = {'num_inference_steps': 100, 'num_images_per_prompt': 2}

generate_image(pipe, prompt, params)
```

```python
# negative_prompt
params = {'num_inference_steps': 100, 'num_images_per_prompt': 2, 'negative_prompt': 'ugly, distorted, low quality'}

generate_image(pipe, prompt, params)
```
